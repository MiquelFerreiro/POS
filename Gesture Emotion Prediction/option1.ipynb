```python
from feat import Detector

detector = Detector(
    face_model="retinaface",
    emotion_model="resmasknet",
    device="cpu"   # use "cuda" if available
)
```

```python
import pandas as pd

def feat_on_video(video_path, sample_every_n_frames=3):
    df = detector.detect_video(
        video_path,
        skip_frames=60
    )
    return df
```

```python
base_path = "D:/castphys_60/Data/"

Patient = 1

quadrant = "Q1_1"

path = f"{base_path}/Patient_{Patient}/{quadrant}/vid_crop.avi"

df = feat_on_video(path)
```
