```python
import cv2
import numpy as np
from feat import Detector

cap = cv2.VideoCapture(video_path)

aus_all = []
frame_count = 0
skip_every = 3
target_width = 640

while True:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_count % skip_every == 0:
        # Resize
        h, w = frame.shape[:2]
        scale = target_width / w
        new_h = int(h * scale)
        frame_resized = cv2.resize(frame, (target_width, new_h))

        # Pass frame directly (no saving)
        results = detector.detect_image(frame_resized)
        aus_all.append(results['aus'].flatten())

    frame_count += 1

cap.release()
aus_all = np.array(aus_all)
```

```python
cap = cv2.VideoCapture(video_path)
aus_all = []
frame_count = 0
skip_every = 3  # adjust to reduce computation
target_width = 640  # e.g., reduce HD to 640x? to speed up

while True:
    ret, frame = cap.read()
    if not ret:
        break
    if frame_count % skip_every == 0:
        # Resize frame to speed up detection
        h, w = frame.shape[:2]
        scale = target_width / w
        new_h = int(h * scale)
        frame_resized = cv2.resize(frame, (target_width, new_h))
        
        results = detector.detect_image(frame_resized)
        aus_all.append(results["aus"].flatten())
    frame_count += 1

cap.release()
aus_all = np.array(aus_all)
```
