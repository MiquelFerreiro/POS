```python
import subprocess
import os

base_path = r"C:\Users\mique\OneDrive\Dokumenty\GitHub\OpenFace_2.2.0_win_x64"

openface_path = rf"{base_path}\FeatureExtraction.exe"

video_path = r"D:\castphys_60\Data\Patient_1\Q1_1\vid_crop.avi"

output_dir = os.path.join(base_path, "Results")

os.makedirs(output_dir, exist_ok=True)

cmd = [
    openface_path,
    "-f", video_path,
    "-out_dir", output_dir,
    "-simscale", "0.5",
    "-feature_interval", "5"
]

result = subprocess.run(cmd, capture_output=True, text=True)

print("Return code:", result.returncode)
print("STDOUT:\n", result.stdout)
print("STDERR:\n", result.stderr)
```
