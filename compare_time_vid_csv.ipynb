```python
import cv2
import pandas as pd
import os

# -------- CONFIG --------
base_path = "d:/castphys_60/Data/"
video_name = "vid_crop.avi"
csv_name = "bio.csv"
elicitation_vid_path =  "d:/castphys_60/video_elicitation/"
paths = [
    "Q1_1", 
    "Q1_2",
    "Q2_1",
    "Q2_2",
    "Q3_1",
    "Q3_2",
    "Q4_1",
    "Q4_2",
    "Q5_1",
    "Q5_2",
    "Q6_1",
    "Q6_2",
    "Q7_1",
    "Q7_2",
    "Q8_1",
    "Q8_2",
    "Q9_1",
    "Q9_2"
]
# ------------------------

print(f"{'Video':12s} {'Video (s)':>12s} {'CSV (s)':>12s} {'    Elicitation Vid (s)':>15s}  {'  Difference (s)':>15s}")
print("-" * 80)

len_vid_elicitation = []

files = os.listdir(elicitation_vid_path)

idx = 0

jdx = 0

#wrong = [8, 9, 13, 15]

while idx < 18:
    
    cap = cv2.VideoCapture(elicitation_vid_path + files[jdx])
    fps = cap.get(cv2.CAP_PROP_FPS)
    frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)
    cap.release()

    video_len = None

    if fps > 0:
        video_len = frames / fps


    jdx = jdx + 1
    len_vid_elicitation.append(video_len)

    
    idx = idx + 1
        

   

for idx in range(len(paths)):

    patient = "Patient_59"

    folder_path = base_path + patient + "/" + paths[idx]

    if not os.path.isdir(folder_path):
        print(f"Error A: {folder_path}")
        continue

    video_path = os.path.join(folder_path, video_name)
    csv_path   = os.path.join(folder_path, csv_name)

    video_len = None
    csv_len   = None

    # ---- Video length ----
    if os.path.exists(video_path):
        cap = cv2.VideoCapture(video_path)
        fps = cap.get(cv2.CAP_PROP_FPS)
        frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)
        cap.release()

        if fps > 0:
            video_len = frames / fps

    # ---- CSV length ----
    if os.path.exists(csv_path):
        df = pd.read_csv(csv_path)
        df[['date', 'clock']] = df['time'].str.split(expand=True)
        t = pd.to_datetime(df['date'] + " " + df['clock'])
        csv_len = (t.iloc[-1] - t.iloc[0]).total_seconds()


        if fps > 0:
            video_len = frames / fps

    # ---- Print ----
    v_str = f"{video_len:.2f}s" if video_len is not None else "—"
    c_str = f"{csv_len:.2f}s"   if csv_len is not None else "—"
    a_str = f"{len_vid_elicitation[idx]:.2f}s"   if len_vid_elicitation[idx] != 0 else "—"

    if (a_str != "—"):
        diff = csv_len - len_vid_elicitation[idx]
    else:
        diff = None
    diff_str = f"{diff:.2f}s"  if diff != None else "—"


    print(f"{paths[idx]:12s} {v_str:>12s} {c_str:>12s} {a_str:>15s}     {diff_str:>15s}")
```
