# Visualizador de dos videos lado a lado con selector
Rellena las rutas `video1_path` y `video2_path` con tus archivos.

```python
from IPython.display import HTML, display
import ipywidgets as widgets

# --- RUTAS A COMPLETAR ---
video1_path = 'RUTA/AL/VIDEO1.mp4'
video2_path = 'RUTA/AL/VIDEO2.mp4'

def video_html(path, width=400):
    return f"""
    <video width='{width}' controls>
      <source src='{path}' type='video/mp4'>
    </video>
    """

selector_left = widgets.Dropdown(options={'Video 1': video1_path, 'Video 2': video2_path}, description='Izquierda:')
selector_right = widgets.Dropdown(options={'Video 1': video1_path, 'Video 2': video2_path}, description='Derecha:')

output = widgets.Output()

def update(_=None):
    with output:
        output.clear_output()
        html = f"<div style='display:flex; gap:20px;'>" \
               f"<div>{video_html(selector_left.value)}</div>" \
               f"<div>{video_html(selector_right.value)}</div>" \
               f"</div>"
        display(HTML(html))

selector_left.observe(update, names='value')
selector_right.observe(update, names='value')

display(widgets.HBox([selector_left, selector_right]))
display(output)

update()
```
