```python
label_map = {
    "Q1": "Q1: â†‘Arousal â†“Val",
    "Q2": "Q2: â†‘Arousal -Val",
    "Q3": "Q3: â†‘Arousal â†‘Val",
    "Q4": "Q4: -Arousal â†“Val",
    "Q5": "Q5: -Arousal -Val",
    "Q6": "Q6: -Arousal â†‘Val",
    "Q7": "Q7: â†“Arousal â†“Val",
    "Q8": "Q8: â†“Arousal -Val",
    "Q9": "Q9: â†“Arousal â†‘Val",
}

def get_label(quadrant):
    q = quadrant.split("_")[0]  # "Q3_2" â†’ "Q3"
    return label_map[q]

```

```python
import numpy as np

X = []
y = []

for bvp in BVPs:
    if bvp is None or bvp.features == []:
        print("Error: Patient", bvp.patient, bvp.quadrant)
        continue

    feat_values = list(bvp.features.values())
    X.append(feat_values)
    y.append(get_label(bvp.quadrant))
    

X = np.array(X)
y = np.array(y)

print(X.shape, y.shape)
print(np.unique(y, return_counts=True))
print("Example of data: ", X[0])
```

LEARNING

```python
# Random Split

type = "Random"

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    stratify=y,      # important for emotions
    random_state=42
)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

print("Example of data: ", X_train[0])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", RandomForestClassifier(
        n_estimators=200,
        random_state=42,
        class_weight="balanced"
    ))
])

```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score

X = []
y = []
groups = []

for bvp in BVPs:
    if bvp is None or bvp.features == []:
        print(f"Skipping Patient_{bvp.patient} {bvp.quadrant}")
        continue

    X.append(list(bvp.features.values()))
    y.append(get_label(bvp.quadrant))
    groups.append(bvp.patient)

X = np.array(X)
y = np.array(y)
groups = np.array(groups)
```

```python
from sklearn.model_selection import StratifiedGroupKFold, cross_val_score

cv = StratifiedGroupKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

scores = cross_val_score(
    pipe,
    X,
    y,
    groups=groups,
    cv=cv,
    scoring="f1_macro"
)

print("F1 macro:", scores.mean(), "Â±", scores.std())

results.append([full_vid_size, window_size, scores.mean(), scores.std()])
```
